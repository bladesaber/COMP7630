```python
def distance_dict(user_rating, movie, user1, user2, method, user_cross_dict):
    user1_rating_list = []
    user2_rating_list = []
    for tid in user_rating[user1].keys():
        if movie!=None:
            if tid in user_rating[user2] and movie!=tid:
                user1_rating_list.append(user_rating[user1][tid])
                user2_rating_list.append(user_rating[user2][tid])
        else:
            if tid in user_rating[user2]:
                user1_rating_list.append(user_rating[user1][tid])
                user2_rating_list.append(user_rating[user2][tid])

    user1_rating_list = np.array(user1_rating_list)
    user2_rating_list = np.array(user2_rating_list)

    if method == 'Euc':
        distance = np.sqrt(np.sum(np.power(user1_rating_list-user2_rating_list,2)))
        similary = 1/(1+distance)
    elif method == 'Person':
        similary =  np.corrcoef(user1_rating_list, user2_rating_list)
    elif method == 'Cosin':
        x1 = user1_rating_list
        x2 = user2_rating_list
        similary =  np.sum(x1*x2)/(np.sqrt(np.sum(x1*x1))*np.sqrt(np.sum(x2*x2)))
    else:
        similary = None
    
    return similary

def user_distance_dict(user_rating, movie, user1, user2, user_cross_dict, method, check=False):
    if user2 in user_cross_dict[user1].keys():
        if check:
            if movie not in user_rating[user2].keys():
                return None
        
        if len(user_rating[user1].keys())<len(user_rating[user2].keys()):
            return distance_dict(user_rating, movie, user1, user2, method=method, user_cross_dict=user_cross_dict)
        else:
            return distance_dict(user_rating, movie, user2, user1, method=method, user_cross_dict=user_cross_dict)
    
    return np.inf
```

```python
def UserCF_dict_debug_1(user_rating, user, user_cross_dict, movie, user_union_dict, method, k=10):
    candidate_users = list(user_cross_dict[user].keys())
    
    score = []
    adjust_list = []
    effect_candidate = []
    
    for candidate in candidate_users:
        similary = user_distance_dict(user_rating, movie, user, candidate, user_cross_dict, method=method, check=True)
        if similary != None:
            score.append(similary)
            adjust_list.append(user_cross_dict[user][candidate])
            effect_candidate.append(candidate)
    
    score = np.array(score)
    adjust_list = np.array(adjust_list)
    effect_candidate = np.array(effect_candidate)
    
    filter_index = adjust_list>5
    score = score[filter_index]
    adjust_list = adjust_list[filter_index]
    effect_candidate = effect_candidate[filter_index]
    
    score_adjust = score
    
    if len(score_adjust)>0:
        argindex = np.argsort(score_adjust)[-k:]

        user_mean_list = []
        for tid in user_rating[user]:
            if tid!=movie:
                user_mean_list.append(user_rating[user][tid])
        user_mean = np.mean(user_mean_list)

        sum_of_score = 0
        sum_of_similart = 0
        
        sum_of_cross = 0
        sum_of_union = 0
        sum_of_precentage_cross = 0
        
        for index in argindex:
            user2 = effect_candidate[index]

            user2_mean_list = []
            for tid in user_rating[user2]:
                user2_mean_list.append(user_rating[user2][tid])
            user2_mean = np.mean(user2_mean_list)
            user2_rating = user_rating[user2][movie]

            sum_of_score += (user2_rating-user2_mean) * score_adjust[index]
            sum_of_similart += score_adjust[index]
                        
            sum_of_cross += user_cross_dict[user][user2]
            sum_of_union += user_union_dict[user][user2]
            sum_of_precentage_cross += user_cross_dict[user][user2]/len(user_rating[user2])
            sum_of_precentage_cross += user_cross_dict[user][user2]/len(user_rating[user])
        
        similary_list = score[argindex]
        predict_adjust = np.maximum(user_mean + sum_of_score/sum_of_similart, 0.)
        predict_adjust = np.minimum(predict_adjust, 10.)
        
        return [predict_adjust, 
                sum_of_cross/len(argindex), sum_of_union/len(argindex), 
                sum_of_precentage_cross/len(argindex), len(argindex),
               np.max(similary_list), np.mean(similary_list[-3:]),
               np.mean(similary_list[-5:]), np.mean(similary_list)]
    else:
#         user_mean_list = []
#         for tid in user_rating[user]:
#             if tid!=movie:
#                 user_mean_list.append(user_rating[user][tid])
#         user_mean = np.mean(user_mean_list)
#         return user_mean
        return None
```

```python
def evalute_single_user_debug_1(user, user_rating, user_cross_dict, user_union_dict, method):
    predict_scores = []
    
    for movie in user_rating[user].keys():
        pack_result = UserCF_dict_debug_1(user_rating, user,user_cross_dict, movie,method=method, k=10,user_union_dict=user_union_dict)
        if pack_result != None:
            predict_score, sum_of_cross, sum_of_union, sum_of_precentage_cross, number_score, max_s, mean_3, mean_5, mean_s = pack_result
            mean_score_list = []
            for tid in user_rating[user]:
                if tid != movie:
                    mean_score_list.append(user_rating[user][tid])
            mean_score = np.mean(mean_score_list)
            
            real_score = user_rating[user][movie]
            predict_scores.append([user, movie, predict_score, mean_score, real_score, sum_of_cross, sum_of_union, 
                                   sum_of_precentage_cross, number_score, max_s, mean_3, mean_5, mean_s])
    return predict_scores
```

```python
def evaluate_mat_process_debug_1(candadite, user_rating, user_cross_dict, user_union_dict, method='Euc'):
    evalution_list = []
    for user in tqdm(candadite):
        predict_scores = evalute_single_user_debug_1(user, user_rating, user_cross_dict, user_union_dict, method)
        evalution_list.extend(predict_scores)
    return pd.DataFrame(evalution_list)
```
