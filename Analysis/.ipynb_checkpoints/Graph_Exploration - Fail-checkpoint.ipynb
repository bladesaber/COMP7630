```python
import pandas as pd
import matplotlib.pyplot as plt
import gc
import pickle
import numpy as np
from tqdm import tqdm
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from torch.autograd import Variable
from scipy.sparse import csc_matrix
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
import csv
import networkx as nx
from networkx.algorithms.community import *
```

```python
df = pd.read_pickle('./train_set.pkl')
```

```python
df.drop(['id'], axis=1, inplace=True)
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
tid_label_encoder = LabelEncoder()
tid_label_encoder.fit(df['tid'])

user_label_encoder = LabelEncoder()
user_label_encoder.fit(df['userName'])

df['tid_enc'] = tid_label_encoder.transform(df['tid'])
df['user_enc'] = user_label_encoder.transform(df['userName'])
```

```python
h = df.nunique()
h
```

```python
rating_index = df['rating'].sort_values().index
rating_1_index = rating_index[:int(df.shape[0]/2)]
rating_2_index = rating_index[int(df.shape[0]/2):]
```

```python
df.loc[rating_1_index, 'rating_classify'] = 1
df.loc[rating_2_index, 'rating_classify'] = 2
```

```python
sns.distplot(df['rating_classify'], hist=True, kde=True, rug=False, bins=2)
```

```python
sub_df = df[df['rating_classify']==2]
```

```python
sub_df.head()
```

```python
user_tid_pack = []
for i in sub_df.groupby(['user_enc']):
    tid = i[1].tid_enc.values
    user_tid_pack.append(tid)
```

```python
te = TransactionEncoder()
te_ary = te.fit(user_tid_pack).transform(user_tid_pack)
te_df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
fp_model = fpgrowth(te_df, min_support=0.02, use_colnames=True, max_len=2)
```

```python
node_set = []
num = 0
for index, row in tqdm(fp_model.iterrows()):
    if len(row.itemsets)==2:
        id_set =  list(row.itemsets)
        from_id = id_set[0]
        to_id = id_set[1]
        node_set.append([from_id, to_id, row.support*100])
        
        num += 1
print(num)
```

```python
select_node_set = np.array(node_set[:200])
select_node_set[:, 2] = select_node_set[:, 2]/np.max(select_node_set[:, 2])

graph = nx.Graph()
for from_id, to_id, weight in tqdm(select_node_set):
    graph.add_edge(from_id, to_id, weight=weight)
```

```python
nx.spring_layout(graph)
nx.draw_networkx(graph, node_size=30, with_labels=False)
```

```python
try:
    import pydot
    from networkx.drawing.nx_pydot import graphviz_layout
except ImportError:
    raise ImportError("This example needs Graphviz and either PyGraphviz or pydot")
```
