```python
import pandas as pd
import matplotlib.pyplot as plt
import gc
import pickle
import numpy as np
import itertools
from tqdm import tqdm
import seaborn as sns
from scipy import stats
```

```python
def item_distance_dict(item_rating, user, movie1, movie2, method):
    movie1_rating_list = []
    movie2_rating_list = []
    for uid in item_rating[movie1].keys():
        if user!=None:
            if uid in item_rating[movie2] and user!=uid:
                movie1_rating_list.append(item_rating[movie1][uid])
                movie2_rating_list.append(item_rating[movie2][uid])
        else:
            if uid in item_rating[movie2]:
                movie1_rating_list.append(item_rating[movie1][uid])
                movie2_rating_list.append(item_rating[movie2][uid])

    movie1_rating_list = np.array(movie1_rating_list)
    movie2_rating_list = np.array(movie2_rating_list)

    if method == 'Euc':
        distance = np.sqrt(np.sum(np.power(movie1_rating_list-movie2_rating_list,2)))
        return 1/(1+distance)
    elif method == 'Person':
        return np.corrcoef(movie1_rating_list, movie2_rating_list)
    elif method == 'Cosin':
        x1 = movie1_rating_list
        x2 = movie2_rating_list
        return np.sum(x1*x2)/(np.sqrt(np.sum(x1*x1))*np.sqrt(np.sum(x2*x2)))
    else:
        return None

def movie_distance_dict(item_rating, user, movie1, movie2, item_cross_dict, method, check=False):
    if movie2 in item_cross_dict[movie1].keys():
        if check:
            if user not in item_rating[movie2].keys():
                return None
        
        if len(item_rating[movie1].keys())<len(item_rating[movie2].keys()):
            return item_distance_dict(item_rating, user, movie1, movie2, method)
        else:
            return item_distance_dict(item_rating, user, movie2, movie1, method)
    
    return np.inf
```

```python
def ItemCF_dict_debug_1(item_rating, movie, item_cross_dict, user, method, k=10):
    candidate_movies = list(item_cross_dict[movie].keys())
    
    score = []
    adjust_list = []
    effect_candidate = []
    
    for candidate in candidate_movies:
        similary = movie_distance_dict(item_rating, user, movie, candidate, item_cross_dict, method=method, check=True)
        if similary != None:
            score.append(similary)
            adjust_list.append(item_cross_dict[movie][candidate])
            effect_candidate.append(candidate)
    
    score = np.array(score)
    adjust_list = np.array(adjust_list)
    effect_candidate = np.array(effect_candidate)
    
    score_adjust = score
    
    if len(score_adjust)>0:
        argindex = np.argsort(score_adjust)[-k:]

        movie_mean_list = []
        for uid in item_rating[movie]:
            if uid!=user:
                movie_mean_list.append(item_rating[movie][uid])
        movie_mean = np.mean(movie_mean_list)

        sum_of_score = 0
        sum_of_similart = 0
        
        sum_of_cross = 0
        sum_of_precentage_cross = 0
        
        for index in argindex:
            movie2 = effect_candidate[index]

            movie2_mean_list = []
            for uid in item_rating[movie2]:
                movie2_mean_list.append(item_rating[movie2][uid])
            movie2_mean = np.mean(movie2_mean_list)
            movie2_rating = item_rating[movie2][uid]

            sum_of_score += (movie2_rating-movie2_mean) * score_adjust[index]
            sum_of_similart += score_adjust[index]
                        
            sum_of_cross += item_cross_dict[movie][movie2]
            sum_of_precentage_cross += item_cross_dict[movie][movie2]/len(item_rating[movie])
            sum_of_precentage_cross += item_cross_dict[movie][movie2]/len(item_rating[movie2])
        
        similary_list = score[argindex]
        predict_adjust = np.maximum(movie_mean + sum_of_score/sum_of_similart, 0.)
        predict_adjust = np.minimum(predict_adjust, 10.)
        
        return [predict_adjust, 
                sum_of_cross/len(argindex), sum_of_precentage_cross/len(argindex)/2, len(argindex),
               np.max(similary_list), np.mean(similary_list[-3:]),
               np.mean(similary_list[-5:]), np.mean(similary_list)]
    else:
        return None
```

```python
def evalute_single_movie_debug_1(movie, item_rating, item_cross_dict, method):
    predict_scores = []
    
    for user in item_rating[movie].keys():
        pack_result = ItemCF_dict_debug_1(item_rating, movie, item_cross_dict, user, method, k=10)
        if pack_result != None:
            predict_score, sum_of_cross, sum_of_precentage_cross, number_score, max_s, mean_3, mean_5, mean_s = pack_result
            mean_score_list = []
            for uid in item_rating[movie]:
                if uid != user:
                    mean_score_list.append(item_rating[movie][uid])
            mean_score = np.mean(mean_score_list)
            
            real_score = item_rating[movie][user]
            predict_scores.append([user, movie, predict_score, mean_score, real_score, sum_of_cross,
                                   sum_of_precentage_cross, number_score, max_s, mean_3, mean_5, mean_s])
    return predict_scores
```

```python
def evaluate_mat_process_debug_1(candadite, item_rating, item_cross_dict, method='Euc'):
    evalution_list = []
    for movie in tqdm(candadite):
        predict_scores = evalute_single_movie_debug_1(movie, item_rating, item_cross_dict, method)
        evalution_list.extend(predict_scores)
    return pd.DataFrame(evalution_list)
```
